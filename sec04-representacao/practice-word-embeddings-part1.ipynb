# *Word Embeddings* Estáticas

## Carregamento

Inicializando as word embeddings pré-treinadas
> `'pt'` para indicar que o idioma é Português

O modelo *fastText* retorna, além de *word embeddings*, *embeddings* para *n-grams* de caracteres que também estão no vocabulário.

Fornecer informações sobre sub-palavras possui algumas vantagens: 
1. permite que o modelo saiba que diferentes tokens podem ser variações da mesma palavra 
2. ajuda a representar palavras OOV com base em sua ortografia 
3. manipula erros ortográficos

Os vetores pré-treinados para o Português foram geraos a partir de milhões de *tokens* em Português, resultantes do corpora formado pelos dados do Wikipédia e *Common Crawl*

```python
ftext_embedding = flair.embeddings.WordEmbeddings('pt')
```

```python
print("São embeddings estáticos? = {}".format(ftext_embedding.static_embeddings))
print("Tamanho do vocabulário = {} tokens".format(len(ftext_embedding.vocab)))
print("Dimensão dos embeddings = {}".format(ftext_embedding.embedding_length))
print("Tipos dos embeddings = {}".format(ftext_embedding.embedding_type))

ftext_embedding.get_named_embeddings_dict()
```

```python
ftext_embedding.vocab  
```

## Como Usar

```python
sent = flair.data.Sentence("O gato caçou o rato.")

sent.tokens
```

```python
ftext_embedding.embed(sent)

sent.get_token(token_id=2).embedding
```

```python
len(sent.get_token(2).embedding)
```

```python
ftext_embedding.vocab["PLN"]
```

```python
ftext_embedding.vocab["pln"]
```

```python
sent_1 = flair.data.Sentence("Pesquisei sobre receitas de bolo de laranja")
sent_2 = flair.data.Sentence("Há um bolo de coisas sobre a mesa")

ftext_embedding.embed([sent_1, sent_2])
```

```python
print("Sentença 1 =\n{}\n\nSentença 2 =\n{}".format('\n'.join(map(str, sent_1.tokens)), 
                                                    '\n'.join(map(str, sent_2.tokens))))
```

```python
sent_1.get_token(5), sent_2.get_token(3)
```

```python
bolo_emb_sent_1 = sent_1.get_token(5).embedding
bolo_emb_sent_2 = sent_2.get_token(3).embedding

print(bolo_emb_sent_1 == bolo_emb_sent_2)
print("\nÍndice da palavra 'bolo' = {}".format(ftext_embedding.vocab["bolo"]))
```

```python
sobre_emb_sent_1 = sent_1.get_token(2).embedding
sobre_emb_sent_2 = sent_2.get_token(6).embedding

print(sobre_emb_sent_1 == sobre_emb_sent_2)
print("\nÍndice da palavra 'sobre' = {}".format(ftext_embedding.vocab["sobre"]))
```
